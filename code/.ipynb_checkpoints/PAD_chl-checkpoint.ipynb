```python
#import packages

#!pip install netcdf4
import xarray as xr
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
pad_2010 = pd.read_csv('pad_water_quality_2010.csv', delimiter = ',', header = 9)
display(pad_2010)
```

```python
pad_2010['Lon'] = pad_2010['Decimal Degrees.1']
pad_2010['Lat'] = pad_2010['Decimal Degrees']
pad_2010['depth'] = pad_2010['m']
pad_2010['chl'] = pad_2010['ug/L']

pad_2010 = pad_2010[['Lon','depth','chl']]
```

```python
pad_2010 = pad_2010.loc[pad_2010['depth']>-9999.0]
print(pad_2010)
```

```python
pad = pad_2010.groupby(['depth', 'Lon'], dropna = False).mean()
pad = pad.to_xarray()
```

```python
fig = plt.figure(figsize = (10,5))
Lon = pad_2010['Lon'].values
depth = pad_2010['depth'].values

ax = plt.gca()
fig1 = plt.scatter(Lon, depth, s= 100, c= pad_2010['chl'], marker= 'o', ls='-', lw= 2)
c = plt.colorbar(fig1, ax=ax)
c.set_label('chlorophyll concentration (ug/L)', fontsize= 15)
plt.xlabel('Longitude', fontsize= 15)
plt.ylabel('Depth (m)', fontsize= 15)
plt.title('')
```
