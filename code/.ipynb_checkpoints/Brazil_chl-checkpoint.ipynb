```python
#import packages

!pip install netcdf4
import xarray as xr
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy import interpolate
```

```python
brazil = pd.read_csv('Field_measurements_Lago_Curuai.csv', delimiter = ',', header = 29)
display(brazil)

#index data
lon = brazil['Longitude']
lat = brazil['Latitude']
temp = brazil['Temperature']
#chl = brazil['Chlorophyll_a']

chl = brazil.loc[brazil['Chlorophyll_a']<1500.0]
```

```python
fig, (ax, ax2) = plt.subplots(1,2)
ax.scatter(chl['Longitude'], chl['Latitude'], c= chl['Chlorophyll_a'])
c = plt.colorbar(ax=ax2)
```

```python
# make plot
fig = plt.figure(figsize = (10,5))

ax = plt.gca()
fig2 = plt.scatter(chl['Longitude'], chl['Latitude'], c= np.log(chl['Chlorophyll_a']))
# fig2 = plt.scatter(lon, lat, s= 50, c= chl, cmap = 'viridis', marker= 'o', ls='-', lw= 1, edgecolor= 'black')
c = plt.colorbar(fig2, ax=ax)
c.set_label('Chlorophyll Concentration (ug/L)', fontsize= 15)
plt.xlabel('Longitude', fontsize= 15)
plt.ylabel('Latitude (m)', fontsize= 15)
plt.title('Chlorophyll Concentration at Lago Curuai, Para, Brazil (2003-2004)', fontsize = 15)
plt.grid()
#plt.savefig('Brazil_chl.png')
```
